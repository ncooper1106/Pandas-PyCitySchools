```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Preview the combined data
school_data_complete = school_data_complete.sort_values(["school_name"])
school_data_complete = school_data_complete.reset_index(drop=True)
school_data_complete = school_data_complete.rename(columns={"grade": "Grade"})
school_data_complete.head()
```

```python
# Rename school_name and type columns
school_data_complete = school_data_complete.rename(columns={"school_name": "School", "type" :'Type'})

# Group the data by school and type
school_type_grouped = school_data_complete.groupby(['School', 'Type'])
school_type = school_type_grouped.mean()
school_type.head()
```

```python
# Calculate number of schools
num_schools=len(school_data_complete.School.unique())
num_schools
```

```python
# Calculate number of Total Students
num_students = school_type['size'].sum()
num_students
```

```python
# Calculate Total Budget
total_budget = school_type['budget'].sum()
total_budget
```

```python
# Calculate Average Math Score
avg_math = school_data_complete['math_score'].mean()
avg_math
```

```python
# Calculate Average Reading Score
avg_read = school_data_complete['reading_score'].mean()
avg_read
```

```python
# Calculate % Passing Math and Reading
school_data_complete['Passing Math Index'] = ''

# Set up passing math index
for i in range(0, len(school_data_complete)):
    if school_data_complete.iloc[i,6] < 70:
        school_data_complete.iloc[i,11] = 0
    else:
        school_data_complete.iloc[i,11] = 1

# Set up passing reading index 
school_data_complete['Passing Reading Index'] = ''

for i in range(0, len(school_data_complete)):
    if school_data_complete.iloc[i,5] < 70:
        school_data_complete.iloc[i,12] = 0
    else:
        school_data_complete.iloc[i,12] = 1
        
school_data_complete.head()
```

```python
# Calculate % Passing Math
pass_math = school_data_complete['Passing Math Index'].sum() / num_students * 100
pass_math
```

```python
# Calculate % Passing Reading
pass_read = school_data_complete['Passing Reading Index'].sum() / num_students * 100
pass_read
```

```python
# Calculate % Overall Passing Rate
ovr_pass = (avg_math + avg_read) / 2
ovr_pass
```

```python
# Create District Summary Table
district = {"Total Schools":[num_schools], "Total Students":[num_students], "Total Budget":[total_budget], "Average Math Score":[avg_math], "Average Reading Score":[avg_read], "% Passing Math":[pass_math], "% Passing Reading": [pass_read], "% Overall Passing Rate": [ovr_pass]}
df_district = pd.DataFrame(district)

#Format Datafram
df_district['Total Students']=df_district['Total Students'].map("{:,}".format)
df_district['Total Budget']=df_district['Total Budget'].map("${:,}".format)

df_district
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math

  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Create an overview table that summarizes key metrics about each school
summary_table = school_type.drop(['Student ID', 'School ID'], axis=1)
summary_table.head()
```

```python
# rename columns
summary_table = summary_table.rename(columns={"size":"Total Students", "budget": "Total Budget", "math_score": "Avg Math Score", "reading_score": "Avg Reading Score"})

summary_table.head()
```

```python
# Calculate Per Student Budget
summary_table['Per Student Budget'] = summary_table['Total Budget']/summary_table['Total Students']
summary_table.head()
```

```python
# Calulate % passing math and reading
math_reading = school_data_complete.groupby(['School', 'Type'])

summary_table["Passing Math"] = math_reading['Passing Math Index'].sum() / summary_table['Total Students']*100
summary_table["Passing Reading"] = math_reading['Passing Reading Index'].sum() / summary_table['Total Students']*100
summary_table.head()
```

```python
# Calculate Overall Passing Rate (Average of Passing Math and Passing Reading)
summary_table['Overall Passing Rate'] = (summary_table['Passing Math'] + summary_table['Passing Reading']) / 2
summary_table.head()
```

```python
# Rearrange Columns
summary_table = summary_table[['Total Students', 'Total Budget', 'Per Student Budget', 'Avg Math Score', 'Avg Reading Score', 'Passing Math', 'Passing Reading', 'Overall Passing Rate']]
summary_table.head(10)
```

```python
# Format columns
format_table = summary_table

format_table['Total Students'] = format_table['Total Students'].map("{:,}".format)
format_table['Total Budget'] = format_table['Total Budget'].map("${:,}".format)
format_table['Per Student Budget'] = format_table['Per Student Budget'].map("${:,}".format)
format_table['Avg Math Score'] = format_table['Avg Math Score'].map("{:.2F}".format)
format_table['Avg Reading Score'] = format_table['Avg Reading Score'].map("{:.2F}".format)
format_table['Passing Math'] = format_table['Passing Math'].map("{:.2F}%".format)
format_table['Passing Reading'] = format_table['Passing Reading'].map("{:.2F}%".format)
format_table['Overall Passing Rate'] = format_table['Overall Passing Rate'].map("{:.2F}%".format)
format_table
```
